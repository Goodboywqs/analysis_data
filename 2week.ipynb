```python
def mathcal(n):
    s=[]
    for i in range(100+n):
        s.append(100+i)
    return

def ufunccal(n):
    s=np.array(range(n+1))+100
    return

%timeit mathcal(1000000)
%timeit ufunccal(1000000)
```

2.自定义ufunc函数：用自定义函数对数组元素进行操作，用frompyfun函数将一个计算单个元素的函数转换成ufunc函数
 frompyfunc(func，n_in，n_out)

```python
def liftscore(n):
    n_new=np.sqrt((n^2)*100)
    return n_new
```

```python
score=np.array([87,77,56,100,60])
score_1=np.frompyfunc(liftscore,1,1)(score)
score_1
```

```python
score_1=score_1.astype(float)  #feompyfunc转换的func函数类型是object 再调用数组的astype方法将其转换为浮点数组
score_1.dtype
```

```python
score_2=np.vectorize(liftscore,otypes=[float])(score)  #vectorize 可以实现类似功能，他可以通过otypes指定返回数组的元素类型
print score_2
print score_2.dtype
print any(score_1==score_2)
```

```python
N=5  #事先定义条形的数目
menMeans=(20,35,30,35,27)
womenMeans=(25,32,34,20,25)
menStd=(2,3,4,1,2)
womenStd=(3,5,2,3,3)
ind=np.arange(N)  #ind表示每个条形的编号
width=0.45

p1=plt.bar(ind,menMeans,width,color='grey',yerr=menStd)
p2=plt.bar(ind,womenMeans,width,color='lightblue',bottom=menMeans,yerr=womenStd)  #参数bottom可以指定堆叠的基准，即堆叠图形下方的变量
plt.show()
```

### 4.3.16等高线图

等高线图常用于地理上的地形描述，在统计数据分析中也可以用来描述数据的分布情形

```python
u=np.linspace(-3,3,30)
x,y=np.meshgrid(u,u)
r=0.6;mux=0;muy=0;sx=1;sy=1;
z=(1/(2*3.1415926897*sx*sy*np.sqrt(1-r*r)))*np.exp((-1/(2*(1-r*r)))*(((x-mux)**2)/(sx**2)-2*r*(x-mux)*(y-muy)/(sx*sy)+((y-muy)**2)/(sx**2)))
plt.contourf(x,y,z,alpha=.35,cmap=cm.gist_earth)
c=plt.contour(x,y,z,colors='black',linewidth=.5)
plt.clabel(c,inline=1,fontsize=10)
plt.show()
```
